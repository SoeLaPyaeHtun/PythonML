```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_pairwise(df):
    sns.set(style='darkgrid')
    sns.pairplot(df, hue='origin',height=1.5, aspect=1.0,palette='tab10')
    plt.show()
```

```python
df = pd.read_csv('auto-mpg.csv')
corr_mat = df.corr()
plot_pairwise(df)
label = 'mpg'
candidates_df = corr_mat.loc[
    (corr_mat[label] < -0.5) | (corr_mat[label] > 0.5),
    (corr_mat[label] < -0.5) | (corr_mat[label] > 0.5)
]

candidates_df = candidates_df.drop(columns=[label])
candidates_df = corr_mat[label]

print(candidates_df)

```
